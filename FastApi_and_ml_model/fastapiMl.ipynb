# feature engineering 

```python
#Bmi  weight /height **2

df["bmi"]=df["weight"]/(df["height"]**2)
```

```python
df.describe()
```

```python
df["occupation"].unique()
```

```python
def age_group(age):
    if age<23:
        return "young"
    elif age < 45:
        return "adult"
    elif age< 60:
        return "middle_a"
    return "senior"
        
```

```python
df["age_group"]=df["age"].apply(age_group)
```

```python
df
```

```python
def life_risks(row):
    if row["smoker"] and row["age"]>30:
        return "high"
    elif row["smoker"] or row["age"] >27:
        return "medium"
    return "low"
```

```python
df["life_risk"]=df.apply(life_risks,axis=1)
```

```python
df["city"].unique()
```

```python
def city_tier(row):
    tier_11=["Delhi","Mumbai","Bangalore","Chennai","Hyderabad","Kolkata","Pune"]
    tier_22=["Chandigarh","Lucknow","Gaya","Mysore","Jalandhar","Kota","Indore","Jaipur"]
    if row in tier_11:
           return 1
    elif row in tier_22:
           return 2
    return 3

```

```python
df["tier_city"]=df["city"].apply(city_tier)
```

```python
df.drop(["age",	"weight",	"height","smoker","city"],axis=1,inplace=True)
```

```python
df
```

```python
x=df.drop(columns=["insurance_premium_category"],axis=1)
```

```python
y=df["insurance_premium_category"]
```

```python
df.info()
```

```python
num_col=x.select_dtypes(include=['int64','float64']).columns
categorical_col=x.select_dtypes(include=["object"]).columns
```

```python
num_col
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline

preprocess=ColumnTransformer(
    transformers=[("cat",OneHotEncoder(),categorical_col),
               ("num","passthrough",num_col)])
```

```python
from sklearn.ensemble import RandomForestClassifier
Pipline_=Pipeline(steps=[("preprosses",preprocess),
                         ("classified",RandomForestClassifier())])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
Pipline_.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

pred=Pipline_.predict(x_test)
```

```python
accuracy_score(y_test,pred)
```
